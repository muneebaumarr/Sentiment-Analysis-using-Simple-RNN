```python
from tensorflow.keras.preprocessing.text import  one_hot
```

```python
sent = ['The Glass mof milk', 'The glass OF Juice', 'The Cup Of Tea', 'I am a good boy', 'I am a good developer', 'Understand the meaning of words', 'your videos are good']
sent
```

```python
# Define the vocabulary size

voc_size = 10000
```

```python
## One Hot Respresentation

one_hot_rep=[one_hot(words, voc_size) for words in sent]
one_hot_rep
```

```python
#word embediing representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length = 8
embedding_docs = pad_sequences(one_hot_rep, padding='pre',maxlen=sent_length)
print(embedding_docs)
```

```python
#Feature Repesentation

dim = 10

model = Sequential()
model.add(Embedding(voc_size, dim,input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedding_docs)
```
